```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from skimage import data
from scipy.ndimage import affine_transform

# Load the camera image from skimage (grayscale image)
image = data.camera()

# Create initial 2x2 identity matrix
A = np.array([[1.0, 0.0],
              [0.0, 1.0]])

# Define transformation function
def apply_transformation(matrix, image):
    # Invert matrix because affine_transform uses the inverse
    try:
        inv_matrix = np.linalg.inv(matrix)
    except np.linalg.LinAlgError:
        inv_matrix = np.eye(2)
    return affine_transform(image, inv_matrix, offset=0, order=1, mode='constant')

# Create plot
fig = plt.figure(figsize=(10, 6))
gs = fig.add_gridspec(3, 4)

# Axes for sliders
ax_slider00 = fig.add_subplot(gs[0, 0])
ax_slider01 = fig.add_subplot(gs[0, 1])
ax_slider10 = fig.add_subplot(gs[1, 0])
ax_slider11 = fig.add_subplot(gs[1, 1])

# Axes for matrix text and image
ax_matrix = fig.add_subplot(gs[:, 2])
ax_image = fig.add_subplot(gs[:, 3])
ax_image.axis('off')

# Sliders setup
slider00 = Slider(ax_slider00, 'A[0,0]', -2.0, 2.0, valinit=1.0)
slider01 = Slider(ax_slider01, 'A[0,1]', -2.0, 2.0, valinit=0.0)
slider10 = Slider(ax_slider10, 'A[1,0]', -2.0, 2.0, valinit=0.0)
slider11 = Slider(ax_slider11, 'A[1,1]', -2.0, 2.0, valinit=1.0)

# Initial matrix display and image
text_matrix = ax_matrix.text(0.1, 0.5, str(A), fontsize=14, verticalalignment='center')
im_display = ax_image.imshow(apply_transformation(A, image), cmap='gray')

# Update function
def update(val):
    A[0, 0] = slider00.val
    A[0, 1] = slider01.val
    A[1, 0] = slider10.val
    A[1, 1] = slider11.val

    transformed = apply_transformation(A, image)
    im_display.set_data(transformed)
    text_matrix.set_text(f"[{A[0,0]:.2f}  {A[0,1]:.2f}]\n[{A[1,0]:.2f}  {A[1,1]:.2f}]")
    fig.canvas.draw_idle()

# Connect sliders
slider00.on_changed(update)
slider01.on_changed(update)
slider10.on_changed(update)
slider11.on_changed(update)

plt.tight_layout()
plt.show()
```
